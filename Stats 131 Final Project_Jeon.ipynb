```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("kc_house_data.csv")
```

```python
data.head()
```

```python
x = data[['bedrooms','bathrooms','sqft_living','sqft_lot','floors','waterfront','view',
'condition','grade','sqft_above','sqft_basement','yr_built','zipcode','lat','long']]
y = data['price']
```

```python
x1 = x.copy()
y1 = y.copy()
```

```python
sub_data = pd.DataFrame(x1)
sub_data['price'] = y1
sub_data['price_million'] = y1/1000000
sub_data.head()
```

```python
sub_data.describe()
```

```python
sns.jointplot(sub_data.lat.values, sub_data.long.values)
plt.ylabel('Longitude')
plt.xlabel('Latitude')
plt.show()
```

```python
sns.lmplot("sqft_living", "price_million", data=sub_data, hue="bedrooms", fit_reg=False)
plt.title("price vs sqft by bed rooms", fontsize=10)
plt.xlabel("Sqft Living")
plt.ylabel("Price(million)")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=104951689)
```

```python
from sklearn import linear_model
regression = linear_model.LinearRegression()
regression.fit(x_train,y_train)
print("Score: ", regression.score(x_test,y_test))
```

```python
predict_0 = regression.predict([[3,1,1180,5650,1,0,0,3,7,1180,0,1955,98178,47.5112,-122.257]])
predict_1 = regression.predict([[3,2.25,2570,7242,2.0,0,0,3,7,2170,400,1951,98125,47.7210,-122.319]])
predict_2 = regression.predict([[2,1.00,770,10000,1.0,0,0,3,6,770,0,1933,98028,47.7379,-122.233]])
predict_3 = regression.predict([[4,3.00,1960,5000,1.0,0,0,5,7,1050,910,1965,98136,47.5208,-122.393]])
predict_4 = regression.predict([[3,2.00,1680,8080,1.0,0,0,3,8,1680,0,1987,98074,47.6168,-122.045]])
print("1st(act vs pred)->", "actual:", y.loc[0], ", predicted:", round(float(predict_0)), "accur: ", round(float(predict_0)/y.loc[0]*100,2), "%")
print("2nd(act vs pred)->", "actual:", y.loc[1], ", predicted:", round(float(predict_1)), "accur: ", round(float(predict_1)/y.loc[1]*100,2), "%")
print("3rd(act vs pred)->", "actual:", y.loc[2], ", predicted:", round(float(predict_2)), "accur: ", round(float(predict_2)/y.loc[2]*100,2), "%")
print("4th(act vs pred)->", "actual:", y.loc[3], ", predicted:", round(float(predict_3)), "accur: ", round(float(predict_3)/y.loc[3]*100,2), "%")
print("5th(act vs pred)->", "actual:", y.loc[4], ", predicted:", round(float(predict_4)), "accur: ", round(float(predict_4)/y.loc[4]*100,2), "%")
```

```python
from sklearn import ensemble
clf = ensemble.GradientBoostingRegressor(n_estimators = 500, max_depth = 5, min_samples_split = 2, learning_rate = 0.1, loss = 'ls')
clf.fit(x_train,y_train)
y_test_predict2 = clf.predict(x_test)
```

```python
predict_0_2 = clf.predict([[3,1,1180,5650,1,0,0,3,7,1180,0,1955,98178,47.5112,-122.257]])
predict_1_2 = clf.predict([[3,2.25,2570,7242,2.0,0,0,3,7,2170,400,1951,98125,47.7210,-122.319]])
predict_2_2 = clf.predict([[2,1.00,770,10000,1.0,0,0,3,6,770,0,1933,98028,47.7379,-122.233]])
predict_3_2 = clf.predict([[4,3.00,1960,5000,1.0,0,0,5,7,1050,910,1965,98136,47.5208,-122.393]])
predict_4_2 = clf.predict([[3,2.00,1680,8080,1.0,0,0,3,8,1680,0,1987,98074,47.6168,-122.045]])
print("1st(act vs pred)->", "actual:", y.loc[0], ", predicted:", round(float(predict_0_2)), "accur: ", round(float(predict_0_2)/y.loc[0]*100, 2), "%")
print("2nd(act vs pred)->", "actual:", y.loc[1], ", predicted:", round(float(predict_1_2)), "accur: ", round(float(predict_1_2)/y.loc[1]*100, 2), "%")
print("3rd(act vs pred)->", "actual:", y.loc[2], ", predicted:", round(float(predict_2_2)), "accur: ", round(float(predict_2_2)/y.loc[2]*100, 2), "%")
print("4th(act vs pred)->", "actual:", y.loc[3], ", predicted:", round(float(predict_3_2)), "accur: ", round(float(predict_3_2)/y.loc[3]*100, 2), "%")
print("5th(act vs pred)->", "actual:", y.loc[4], ", predicted:", round(float(predict_4_2)), "accur: ", round(float(predict_4_2)/y.loc[4]*100, 2), "%")
```

```python
# linear regression
y_test_predict = regression.predict(x_test)
import sklearn.metrics as sm 
print("Score: ", regression.score(x_test,y_test))
print("RMSE: ", np.sqrt(sm.mean_squared_error(y_test, y_test_predict)))

# Gradient Boosting Regression
print("GBR Score: ", clf.score(x_test,y_test))
print("GMR RMSE: ", np.sqrt(sm.mean_squared_error(y_test, y_test_predict2)))
```
